```python
#演習1-6
a,b,c = map(int, input().split())

n_max = a
n_min = a

if n_max <= b:
    n_max = b
elif n_min > b: 
    n_min = b

if n_max <= c:
    mid = n_max
    n_max = c
elif n_min > c:
    mid = n_min
    n_min = c
else:
    mid = c

print(n_min,mid,n_max)
```
