```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def softmax(V, beta):
    V = V / np.max(V)
    P = np.exp(beta*V)/np.sum(np.exp(beta*V))
    return P

def getChoice(V, beta):
    P = softmax(V, beta)
    if np.random.rand() < P[0]:
        print("Choosed A")
        choice = "A"
    else:
        print("Choosed B")
        choice = "B"
    return choice

def RLagent(A, B, r, hist, alpha):
    Anxt, Bnxt = 0, 0
    if hist == 1:
        Anxt = A + alpha*(r - A)
        Bnxt = B
    else:
        Bnxt = B + alpha*(r - B)
        Anxt = A            
    return Anxt, Bnxt
```

```python
t = 20
hist = np.zeros(20,dtype=int)
goukei = 0
alpha = 0.3
beta = 0.1
l = 0.88
sigma = 3
c = 50
A = np.zeros(21)
B = np.zeros(21)

A[0] = np.ceil(20*np.random.rand()+40) # 40~60の間の数
B[0] = np.ceil(20*np.random.rand()+40)

for i in range(t):
    V = [A[i], B[i]]
    choice = getChoice(V, beta)
    if choice == "A":
        hist[i] = 1
        r = np.random.normal(A[i],20)
    elif choice == "B":
        hist[i] = 2
        r = np.random.normal(B[i],20)
    else:
        break  # A, B以外のキーを押したら終了
        
    if r < 0:
        r = 0
    r = int(r)
    goukei += r
    print(f"{r}点当たりました")
    print(f"合計は{goukei}点です")
    print()
    A[i]=l*A[i]+(1-l)*c+np.random.normal(0,sigma)
    B[i]=l*B[i]+(1-l)*c+np.random.normal(0,sigma)
    A[i+1], B[i+1] = RLagent(A[i],B[i],r,hist[i],alpha)  # 次のトライアルに行く前に学習
    
print()
print("+++++++++++++++++")
print("Aの選択数: "+str(np.sum(hist[hist==1]))+", Bの選択回数: "+str(20-np.sum(hist[hist==1])))
print("合計点: "+str(goukei))
print("+++++++++++++++++")
```

```python
plt.plot(A,label="A")
plt.plot(B,label="B")
plt.legend()
plt.show()
```
