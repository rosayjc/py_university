```python
import gym
import numpy as np
```

```python
# 環境を'FrozenLake-v0'にする
env = gym.make("FrozenLake-v0")
n=200

G=np.zeros(n)
for i in range(n):
    state=env.reset()
    while True:
        if i==199:
            env.render()
        action=np.random.choice([0,1,2,3])
        state,reward,done,info=env.step(action)
        
        if done:
            if i==199:
                env.render()
                
            if reward==1:
                print("GOAL!")
                G[i]=1
            else:
                print("落ちた", i+1)
                G[i]=0
            break
S=np.sum(G[-100:])

print()
print("++++++++++++++++")
print(f"最後の100回のgoal確率は{S/100}%です")  
print("++++++++++++++++")
        
```
