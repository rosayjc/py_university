```python
import matplotlib.pyplot as plt
import numpy as np


for beta in range(5,30,5):
    x=np.linspace(-5,5,100)#-5から5までを100等分
    y=1/(1+np.exp(-(beta/10)*x))
    L="beta="+str(beta/10)
    plt.plot(x,y,label=L)
    
G=np.zeros(50)
G=np.hstack((G,np.ones(50)))
plt.plot(x,G,label="greedy",color="black")
plt.legend()
plt.ylabel("% choose(A)")
plt.xlabel("Va-Vb")
plt.show()
```

```python
def softmax(V,beta): #引数はnumpy配列Vとbeta
    V=(V-np.mean(V))/np.std(V) #標準化しておく。しなくてもいいんだけど。
    P=np.exp(beta*V)/np.sum(np.exp(beta*V))#softmaxの式そのまま
    return P #それぞれの選択肢の選択確率の入った配列
```

```python
P=softmax(np.array([65,30]),1)
print(P)
```
