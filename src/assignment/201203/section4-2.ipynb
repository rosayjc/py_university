```python
import numpy as np
import matplotlib.pyplot as plt

x=np.arange(20,100)
y=5*x+np.random.normal(0,100,len(x))#yはxを5倍して、平均0標準偏差30の

plt.scatter(x,y)
```

```python
beta=np.polyfit(x,y,1)
print(beta)
```

```python
yhat=x*beta[0]+beta[1]

plt.scatter(x,y)
plt.plot(x,yhat,color="red")
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x.reshape(-1, 1),y.reshape(-1, 1))
#fit()には、(n行1列)の行列を入れる
print(model.coef_) # coeficient  y=ax+bのa
print(model.intercept_) #intercept, 切片
print(model.score(x.reshape(-1, 1),y.reshape(-1, 1)))
# score()で予測の良さを返します。linearRegression()のモデルではR^2値です
ybar=model.predict(x.reshape(-1,1))# predict()で予測値を返します
plt.scatter(x,y)
plt.plot(x,ybar,color="red")
```
